```python
from matplotlib import pyplot as plt
```

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2,543.3,1075.9,2862.5, 5979.6, 10289.7, 14958.3]
```

```python
#Criando um grafico de linhas
plt.plot(
    years,
    gdp,
    color='green',
    marker = 'o',
    linestyle = 'solid'
)

#Adicionando titulo 
plt.title('Nominal GDP')

#Adicionando rotulo no eixo y 
plt.ylabel('Billions for $')
plt.show()
```

```python
movies = ['Annie Hall', 'Ben-hun', 'casablanca', 'gandhi', 'whest Side Story']
num_oscars = [5,11,3,8,10]
```

```python
#plot as barras com cordenadas x  esquerta [0,1,2...], altura [num oscars]
plt.bar(range(len(movies)), num_oscars)
plt.title('meu filme favorito')
plt.ylabel('# of academy awards')

# rotule o eixo x com os nomes dos filmes nos centros das barras
plt.xticks(range(len(movies)), movies)

plt.show()
```

```python
from collections import Counter
```

```python
grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]

#Agrupe as notas por Deci, mas coloque o 100 como 90

histogram = Counter(min(grade // 10 * 10, 90) for grade in grades)
plt.bar([x + 5 for x in histogram.keys()],histogram.values(),10, edgecolor=(0,0,0))
plt.axis([-5, 105, 0, 5])

plt.xticks([10 * i for i in range(11)])
plt.xlabel('Decile')
plt.ylabel('# for students')
plt.title('Distribuio of exam 1 grades')
plt.show()
    
```
